Imagine que você está tentando descobrir quão parecidas são as preferências de duas pessoas sobre filmes.

A pessoa A dá notas altas para filmes de ação, médios para comédia e baixas para drama.

A pessoa B também dá notas altas para ação, médias para comédia e baixas para drama.

Ambas as pessoas têm um padrão de gosto muito parecido. O que a Similaridade de Cosseno faz é medir o "ângulo" entre os gostos dessas duas pessoas.

Um ângulo pequeno (perto de 0 graus) significa que os gostos são muito parecidos (similaridade alta, perto de 1).

Um ângulo grande (perto de 90 graus) significa que os gostos são muito diferentes (similaridade baixa, perto de 0).

A grande sacada é que a Similaridade de Cosseno ignora a "intensidade" (se as notas são todas 10s ou todas 5s) e se foca apenas no padrão (se o gosto por um tipo de filme sobe ou desce da mesma forma).

Neste script, vamos ver como isso funciona com um exemplo simples.

```python
# =================================================================
# 1. Configuração e Criação dos Dados
# =================================================================

# Importando as bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cosine

print("Passo 1: Criando os 'vetores' de preferências dos usuários.")
print("-" * 50)

# Vetor A representa as notas de 3 produtos dadas pelo Usuário A
# Produto 1, Produto 2, Produto 3
usuario_a = np.array([4, 5, 1])

# Vetor B representa as notas de 3 produtos dadas pelo Usuário B
usuario_b = np.array([3, 4, 1])

# Vetor C representa as notas de 3 produtos de um usuário com gostos diferentes
usuario_c = np.array([1, 1, 5])

print(f"Preferências do Usuário A: {usuario_a}")
print(f"Preferências do Usuário B: {usuario_b}")
print(f"Preferências do Usuário C: {usuario_c}")

# =================================================================
# 2. Visualizando os Vetores (em 2D para ser mais simples)
# =================================================================

print("\nPasso 2: Visualizando o ângulo entre dois vetores (exemplo em 2D).")
print("-" * 50)

# Para visualizar, vamos usar uma versão simplificada com 2 produtos
vetor_a_2d = np.array([4, 5])
vetor_b_2d = np.array([3, 4])

plt.figure(figsize=(8, 8))
plt.quiver(0, 0, vetor_a_2d[0], vetor_a_2d[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Usuário A (2D)', width=0.01)
plt.quiver(0, 0, vetor_b_2d[0], vetor_b_2d[1], angles='xy', scale_units='xy', scale=1, color='red', label='Usuário B (2D)', width=0.01)

plt.xlim(0, 6)
plt.ylim(0, 6)
plt.grid()
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.title('Ângulo entre Vetores')
plt.xlabel('Nota do Produto 1')
plt.ylabel('Nota do Produto 2')
plt.legend()
plt.show()

print("\nOs vetores azuis e vermelhos apontam para a mesma direção geral. O ângulo entre eles é pequeno, indicando alta similaridade.")

# =================================================================
# 3. Calculando a Similaridade de Cosseno
# =================================================================

print("\nPasso 3: Calculando a Similaridade de Cosseno para os dados originais (3D).")
print("-" * 50)

# A Similaridade de Cosseno é 1 - a Distância de Cosseno
# O valor 1 - cosine() nos dá um score de 0 (semelhante) a 1 (idêntico)
# Uma alternativa é usar a função dot product
similaridade_ab = 1 - cosine(usuario_a, usuario_b)
similaridade_ac = 1 - cosine(usuario_a, usuario_c)

print(f"Similaridade de Cosseno entre Usuário A e B: {similaridade_ab:.4f}")
print("O score é bem alto (próximo de 1), o que confirma que os gostos são parecidos.")

print(f"\nSimilaridade de Cosseno entre Usuário A e C: {similaridade_ac:.4f}")
print("O score é bem baixo (próximo de 0), pois os gostos são diferentes.")
```

Em Resumo
A Similaridade de Cosseno mede a similaridade de direção entre dois vetores.

Um score de 1 significa que os vetores apontam na mesma direção (totalmente similar).

Um score de 0 significa que os vetores estão em um ângulo de 90 graus (sem similaridade).

Um score de -1 significa que os vetores apontam em direções opostas.

É uma técnica muito usada em sistemas de recomendação (encontrando usuários com gostos similares) e em processamento de linguagem natural (descobrindo a similaridade de documentos com base na frequência das palavras).
